```python
import easyocr
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

# Step 2: Read in Image


```python
IMAGE_PATH = 'stop-signs-eight-sides.jpg'
```

```python
reader = easyocr.Reader(['en'], gpu = False)
result = reader.readtext(IMAGE_PATH)
result
```

# Step 3: Draw Results


```python
top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])
text = result[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
img = cv2.imread(IMAGE_PATH)
img = cv2.rectangle(img, top_left, bottom_right, (0,255,0), 5)
img = cv2.putText(img, text, top_left, font, 2, (255,255,255),2,cv2.LINE_AA)
plt.imshow(img)
plt.show()
```

# Handling Multiple Lines (English)


```python
IMAGE_PATH = 'ads 2.JPG'
```

```python
reader = easyocr.Reader(['en'], gpu = False)
result = reader.readtext(IMAGE_PATH)
result
```

```python
top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])
text = result[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX

img = cv2.imread(IMAGE_PATH)
for detection in result: 
    top_left = tuple([int(val) for val in detection[0][0]])
    bottom_right = tuple([int(val) for val in detection[0][2]])
    text = detection[1]
    font = cv2.FONT_HERSHEY_SIMPLEX
    
    img = cv2.rectangle(img, top_left, bottom_right, (0,255,0), 5)
    img = cv2.putText(img, text, top_left, font, 1.5, (255,255,255),2,cv2.LINE_AA)


plt.figure(figsize=(10,10))
plt.imshow(img)
plt.show()

```
